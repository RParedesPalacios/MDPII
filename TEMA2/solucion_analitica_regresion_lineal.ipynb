## Solución Analítica para la Regresión Lineal

En este notebook, resolveremos el problema de regresión lineal simple usando una solución analítica basada en la técnica de mínimos cuadrados. La función objetivo es encontrar la recta de la forma:

$ y = wx + b $

que mejor se ajuste a un conjunto de datos de entrenamiento dado.

### Definiendo la Función de Coste

Para medir el ajuste de la recta a los datos, utilizamos la función de coste basada en el error cuadrático medio (Mean Squared Error, MSE):

$$ J(w, b) = \frac{1}{2n} \sum_{i=1}^{n} (y_i - (wx_i + b))^2, $$

donde:
- \\(y_i\\) son los valores reales.
- \\(wx_i + b\\) es la predicción de la recta.
- \\(n\\) es el número de puntos de datos.

### Derivando la Función de Coste

Para encontrar los valores óptimos de los parámetros \\(w\\) y \\(b\\), necesitamos minimizar la función de coste. Esto se logra tomando las derivadas parciales de \\(J(w, b)\\) con respecto a \\(w\\) y \\(b\\) y luego igualándolas a cero.

#### 1. Derivada parcial con respecto a \\(w\\):

La derivada parcial de la función de coste respecto a \\(w\\) es:

$$ \frac{\partial J(w, b)}{\partial w} = -\frac{1}{n} \sum_{i=1}^{n} (y_i - (wx_i + b)) x_i. $$

#### 2. Derivada parcial con respecto a \\(b\\):

La derivada parcial de la función de coste respecto a \\(b\\) es:

$$ \frac{\partial J(w, b)}{\partial b} = -\frac{1}{n} \sum_{i=1}^{n} (y_i - (wx_i + b)). $$

#### Igualando las Derivadas a Cero

Para encontrar los valores óptimos, igualamos estas derivadas a cero:

1. Para \\(w\\):
   
   $$ \sum_{i=1}^{n} (y_i - (wx_i + b)) x_i = 0. $$

2. Para \\(b\\):
   
   $$ \sum_{i=1}^{n} (y_i - (wx_i + b)) = 0. $$

#### Despejando las Fórmulas Óptimas

Podemos resolver estas ecuaciones simultáneamente para encontrar las expresiones óptimas para \\(w\\) y \\(b\\):

Para un conjunto de datos de entrenamiento con puntos $(x_i, y_i)$, la solución analítica para los parámetros óptimos $w$ y $b$ se calcula con las siguientes fórmulas:

#### Pendiente $w$:
$$w = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$

donde $\bar{x}$ y $\bar{y}$ son las medias de los datos de entrada y salida respectivamente.

#### Bias $b$:
$$b = \bar{y} - w \bar{x}$$


Estas fórmulas derivan de la minimización del error cuadrático medio (MSE) de la recta con respecto a los puntos de datos.

### Pasos en el Notebook

1. **Generación de Datos:** Creamos un conjunto de datos sintético donde la relación entre las variables es aproximadamente lineal.
2. **Cálculo de la Solución Analítica:** Usamos las fórmulas de mínimos cuadrados para calcular los valores óptimos de los parámetros \(w\) y \(b\).
3. **Visualización:** Graficamos los datos de entrenamiento junto con la recta óptima obtenida.

### Ejecución

Ejecuta la celda de código a continuación para calcular los parámetros óptimos y visualizar la solución analítica para la regresión lineal.


```python

import numpy as np
import matplotlib.pyplot as plt

# Datos de entrenamiento inventados
np.random.seed(0)  # Para reproducibilidad
x_train = np.linspace(0, 10, 50)  # 50 puntos de entrada entre 0 y 10
y_train = 3 * x_train + 5 + np.random.randn(50) * 2  # y = 3x + 5 + ruido

# Cálculo del valor óptimo analítico
x_mean = np.mean(x_train)
y_mean = np.mean(y_train)

# Cálculo de la pendiente (w)
w_optimal = np.sum((x_train - x_mean) * (y_train - y_mean)) / np.sum((x_train - x_mean) ** 2)

# Cálculo del intercepto (b)
b_optimal = y_mean - w_optimal * x_mean

print(f'Solución analítica: w = {w_optimal:.2f}, b = {b_optimal:.2f}')

# Visualización
plt.scatter(x_train, y_train, color='blue', label='Datos de entrenamiento')
y_pred_analytical = w_optimal * x_train + b_optimal
plt.plot(x_train, y_pred_analytical, color='red', label=f'Recta óptima: y = {w_optimal:.2f}x + {b_optimal:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solución Analítica de la Regresión Lineal')
plt.legend()
plt.show()
```

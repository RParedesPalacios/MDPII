Augmentado de datos en redes neuronales. 

Es bien conocido que el aumentado de datos proporciona capacidad de generalización en las redes neuronales. Esta técnica de aumentado de datos se podría aplicar a nuestro problema de reconocimiento de dígitos (MNIST) mediante MLP.

Se propone como ejercicio implementar el aumentado de datos que se conoce como Mix-UP:

[mixup](https://arxiv.org/pdf/1710.09412)

Un ejemplo de mixup en Keras:

[mixup-keras](https://keras.io/examples/vision/mixup/)

En este ejemplo en concreto se crea un nuevo dataset aumentado, sin embargo lo más adecuado es crear un DataGenerator que pueda generar nuevas muestras "al vuelo".

En la implementación se deben seguir empleando redes MLP (no Convolucionales).

A partir del ejemplo que se muestra a continuación implementar la técnica propuesta. 



```python
from tensorflow import keras
from keras.datasets import mnist
from keras import Sequential
from keras.layers import Dense, Input
from keras.optimizers import SGD


# cargar datos y preprocesar
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print('training set', x_train.shape)
print('test set', x_test.shape)

x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train /= 255
x_test /= 255

num_classes=10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)



# crear datagenerator con mixup

## ....


# crear modelo
model = Sequential()
model.add(Input((784,)))
model.add(Dense(512, activation='relu'))
# ... añadir capas
model.add(Dense(num_classes, activation='softmax'))

model.summary()

sgd=SGD(learning_rate=0.01, momentum=0.9)

model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])



# Entrenar con el generador
batch_size=32
epochs=25

# history = model.fit(...)


# Evaluar con el test set
score = model.evaluate(x_test, y_test, verbose=0)
```
